```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gdf = pd.read_csv(r'C:\Users\ugur\Desktop\nba.games.stats.csv')
gdf.columns
```

```python
gdf['WINorLOSS'] = gdf['WINorLOSS'].replace(['W', 'L'], [1,0]).astype(str).astype(int)
gdf['Home'] = gdf['Home'].replace(['Home', 'Away'], [1,0]).astype(str).astype(int)
gdf['Month'] = [int(m[5:7]) for m in gdf['Date']]
gdf['Year'] = [int(y[:4]) for y in gdf['Date']]

def seasons(d):
  m=d[5:7]
  y=d[:4] 
  if (y =='2014'):
    s='one'
  elif (y=='2015'):
    s='two'
  elif (y=='2016'):
    s='three'
  else:
    s='four'
  return (s)
gdf['Season']=gdf['Date'].apply(seasons)
```

```python
gdf.drop('Date', inplace=True, axis=1)
```

```python
# Calculated average EFF

# `(PTS + REB + AST + STL + BLK 
# âˆ’ ((FGA âˆ’ FGM) + (FTA âˆ’ FTM) + TO))`
```

```python
average_eff_Lakers = ((
    Lakersdata['TeamPoints'].mean() + 
    Lakersdata['Assists'].mean() + 
    Lakersdata['Steals'].mean() + 
    Lakersdata['Blocks'].mean() + 
    Lakersdata['TotalRebounds'].mean() + 
        Lakersdata['FieldGoals'].mean() +
        Lakersdata['FreeThrows'].mean()
    ) - (
        Lakersdata['OpponentPoints'].mean() +
        Lakersdata['Opp.Assists'].mean() +
        Lakersdata['Opp.Steals'].mean() +
        Lakersdata['Opp.Blocks'].mean() +
        Lakersdata['Opp.TotalRebounds'].mean() + 
        Lakersdata['Opp.FreeThrows'].mean() +
        Lakersdata['Opp.FieldGoals'].mean() ))
print(average_eff_Lakers)
```
